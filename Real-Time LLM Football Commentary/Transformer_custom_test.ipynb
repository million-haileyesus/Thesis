```python
# Define hyperparameters
transformer_params = {
    "input_size": X_1_scaled.shape[1],
    "d_model" : 256,
    "nhead" : 8,
    "num_encoder_layers" : 6,
    "num_decoder_layers" : 0,
    "dim_feedforward" : 256,
    "num_classes" : np.unique(y_1_encoded).shape[0],
    "dropout_rate": 0.05,
}
input_size = X_1_scaled.shape[1]
sequence_length = 50
learning_rate = 0.001
batch_size = 256
num_epochs = 75

transformer_model = get_model(C_Transformer, params=transformer_params).to(device)
transformer_model
```

```python
class ToTensorSequential:
    def __call__(self, sample):
        X, dist, y = sample["X"], sample["dist"], sample["y"]
        dist = dist + 1e-9
        return {
            "X": torch.tensor(X, dtype=torch.float32),
            "dist": torch.tensor(dist, dtype=torch.float32),
            "y": torch.tensor(y, dtype=torch.long) 
        }

transform = ToTensorSequential()

class SequenceDataset(Dataset):
    def __init__(self, X, dist, y, transform=None):
        self.X = X
        self.dist = dist # Store numpy arrays
        self.y = y
        self.transform = transform

    def __len__(self):
        return len(self.X) # Number of sequences

    def __getitem__(self, idx):
        # Get the i-th sequence
        sample = {"X": self.X[idx], "dist": self.dist[idx], "y": self.y[idx]}
        if self.transform:
            sample = self.transform(sample) # Convert to tensors + add epsilon

        return sample["X"], sample["dist"], sample["y"]
```

```python
X_comb = np.concatenate([X_1_scaled, X_2_scaled])
y_comb = np.concatenate((y_1_encoded, y_2_encoded))

X_comb_reshaped = X_comb.reshape(-1, sequence_length, input_size)
y_comb_reshaped = y_comb.reshape(-1, sequence_length)

X_3_reshaped = X_3_scaled.reshape(-1, sequence_length, input_size)
y_3_reshaped = y_3_encoded.reshape(-1, sequence_length) 

train_dist_reshaped = train_dist.values.reshape(-1, sequence_length, train_dist.shape[1])
val_dist_reshaped = val_dist.values.reshape(-1, sequence_length, val_dist.shape[1])

train_dataset = SequenceDataset(X_comb_reshaped, train_dist_reshaped, y_comb_reshaped, transform)
validation_dataset = SequenceDataset(X_3_reshaped, val_dist_reshaped, y_3_reshaped, transform)

train_loader = DataLoader(
    dataset=train_dataset, 
    batch_size=batch_size, 
    shuffle=False,
)

validation_loader = DataLoader(
    dataset=validation_dataset, 
    batch_size=batch_size, 
    shuffle=False,
)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(transformer_model.parameters(), lr=learning_rate)

NUM_EPOCHS = 10
for epoch in range(NUM_EPOCHS):
    transformer_model.train() # Set model to training mode
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for i, (src_batch, dist_batch, labels_batch) in enumerate(train_loader):
        # Move data to device
        src_batch = src_batch.to(device)
        dist_batch = dist_batch.to(device)
        labels_batch = labels_batch.to(device)

        # Zero gradients
        optimizer.zero_grad()

        # Forward pass
        # Output shape: [batch_size, seq_len (players), num_classes]
        outputs = transformer_model(src=src_batch, distances=dist_batch)

        # Aggregate output for classification (e.g., mean pooling over players)
        # Shape: [batch_size, num_classes]
        aggregated_output = outputs.reshape(-1, outputs.size(2))
        label_flat = labels_batch.reshape(-1)
        
        # Calculate loss
        loss = criterion(aggregated_output, label_flat)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        # Statistics
        running_loss += loss.item()
        _, predicted_train = torch.max(aggregated_output.data, 1)
        total_train += label_flat.size(0)
        correct_train += (predicted_train == label_flat).sum().item()

        # print(f'Epoch [{epoch+1}/{NUM_EPOCHS}], Batch [{i+1}/{len(train_loader)}], Loss: {loss.item():.4f}')

    epoch_loss = running_loss / len(train_loader)
    epoch_acc_train = 100 * correct_train / total_train
    print(f'Epoch [{epoch+1}/{NUM_EPOCHS}] Train Loss: {epoch_loss:.4f}, Train Accuracy: {epoch_acc_train:.2f}%')

    # --- Validation Loop ---
    transformer_model.eval() # Set model to evaluation mode
    val_loss = 0.0
    correct_val = 0
    total_val = 0
    with torch.no_grad(): # No need to track gradients during validation
        for src_batch, dist_batch, labels_batch in validation_loader:
            src_batch = src_batch.to(device)
            dist_batch = dist_batch.to(device)
            labels_batch = labels_batch.to(device)

            outputs = transformer_model(src=src_batch, distances=dist_batch)
            aggregated_output = outputs.reshape(-1, outputs.size(2))
            label_flat = labels_batch.reshape(-1)
            
            # Calculate loss
            loss = criterion(aggregated_output, label_flat)

            val_loss += loss.item()
            _, predicted_val = torch.max(aggregated_output.data, 1)
            total_val += label_flat.size(0)
            correct_val += (predicted_val == label_flat).sum().item()

    epoch_loss_val = val_loss / len(validation_loader)
    epoch_acc_val = 100 * correct_val / total_val
    print(f'Epoch [{epoch+1}/{NUM_EPOCHS}] Val Loss: {epoch_loss_val:.4f}, Val Accuracy: {epoch_acc_val:.2f}%')

    # Optional: Update learning rate scheduler
    # scheduler.step()

print('Finished Training')
```

```python
def train_one_epoch(model, train_loader, optimizer, criterion, device, is_rnn, is_gnn, is_transformer):
    model.train()
    train_loss = 0.0
    train_acc = 0
    
    train_preds = []
    train_labels = []

    train_dataset_len = len(train_loader.dataset)

    
    for batch_idx, (data, label) in enumerate(train_loader):
        data, label = data.to(device), label.to(device)
        optimizer.zero_grad()

        # Special handling for transformer models
        if is_transformer:
            # Check if it's an encoder-only or encoder-decoder transformer
            is_encoder_only = hasattr(model, 'encoder_only') and model.encoder_only
            
            if is_encoder_only:
                # Encoder-only case - simpler forward pass
                start_idx = batch_idx * batch_size * sequence_length
                end_idx = min(start_idx + batch_size * sequence_length, len(train_dist))
                
                # Extract the relevant slice for this batch
                batch_distances_df = train_dist.iloc[start_idx:end_idx]
                
                # Prepare the distance matrix for this specific batch
                train_distance_matrix = prepare_distance_matrix(
                    batch_distances_df, 
                    batch_size=data.size(0),
                    seq_length=sequence_length
                )
                output = model(data, distances=train_distance_matrix)
            else:
                # Encoder-decoder transformer with teacher forcing
                num_classes = model.decoder.num_classes
                label_onehot = F.one_hot(label, num_classes=num_classes).float()
                output = model(data, label_onehot, teacher_forcing_ratio=0.5)
            
            # Reshape for loss calculation
            output_flat = output.reshape(-1, output.size(2))
            label_flat = label.reshape(-1)
            loss = criterion(output_flat, label_flat)
            
            # Get predictions from the output
            _, pred = torch.max(output, 2)
        else:
            output = model(data)
            if is_rnn:
                output = output.permute(0, 2, 1)
            
            loss = criterion(output, label)
            _, pred = torch.max(output, 1)

        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        train_acc += (pred == label).float().sum().item()

        train_preds.extend(pred.cpu().numpy().flatten())
        train_labels.extend(label.cpu().numpy().flatten())
        
    metrics = {
        "train_labels": np.array(train_preds),
        "train_preds": np.array(train_labels)
    }

    epoch_loss = train_loss / len(train_loader)
    if is_rnn or is_transformer:
        # sequence_length = data.size(1)
        epoch_accuracy = train_acc / (train_dataset_len * sequence_length)
    else:
        epoch_accuracy = train_acc / train_dataset_len

    return epoch_loss, epoch_accuracy, metrics


def validate_one_epoch(model, validation_loader, criterion, device, is_rnn, is_gnn, is_transformer):
    model.eval()
    val_loss = 0.0
    val_acc = 0

    val_preds = []
    val_labels = []

    val_dataset_len = len(validation_loader.dataset)

    with torch.no_grad():
        for batch_idx, (data, label) in enumerate(validation_loader):
            data, label = data.to(device), label.to(device)
            
            # Special handling for transformer models
            if is_transformer:
                # During validation, don't use teacher forcing
                start_idx = batch_idx * batch_size * sequence_length
                end_idx = min(start_idx + batch_size * sequence_length, len(val_dist))
                
                # Extract the relevant slice for this batch
                batch_distances_df = val_dist.iloc[start_idx:end_idx]
                
                # Prepare the distance matrix for this specific batch
                val_distance_matrix = prepare_distance_matrix(
                    batch_distances_df, 
                    batch_size=data.size(0),
                    seq_length=sequence_length
                )
                output = model(data, distances=val_distance_matrix, teacher_forcing_ratio=0.0)
                
                # Reshape for loss calculation
                output_flat = output.reshape(-1, output.size(2))
                label_flat = label.reshape(-1)
                loss = criterion(output_flat, label_flat)
                
                # Get predictions
                _, pred = torch.max(output, 2)
            
            val_loss += loss.item()
            val_acc += (pred == label).float().sum().item()

            val_preds.extend(pred.cpu().numpy().flatten())
            val_labels.extend(label.cpu().numpy().flatten())
    
    epoch_loss = val_loss / len(validation_loader) 
    if is_rnn or is_transformer:
        # sequence_length = data.size(1)
        epoch_accuracy = val_acc / (val_dataset_len * sequence_length)
    else:
        epoch_accuracy = val_acc / val_dataset_len

    metrics = {
        # "precision": precision_score(val_labels, val_preds, average="weighted", zero_division=0),
        # "recall": recall_score(val_labels, val_preds, average="weighted", zero_division=0),
        # "f1": f1_score(val_labels, val_preds, average="weighted", zero_division=0),
        # "val_labels": np.array(val_labels),
        # "val_preds": np.array(val_preds)
    }

    return epoch_loss, epoch_accuracy, metrics
```

```python
optimizer = optim.Adam(transformer_model.parameters(), lr=learning_rate, weight_decay=1e-4)
criterion = nn.CrossEntropyLoss()
for epoch in range(1, num_epochs + 1):
    train_loss, train_acc, train_metrics = train_one_epoch(transformer_model, train_loader, optimizer, criterion=criterion, device=device, is_gnn=False, is_rnn=False, is_transformer=True)
    val_loss, val_acc, val_metrics = validate_one_epoch(transformer_model, validation_loader, criterion=criterion, device=device, is_gnn=False, is_rnn=False, is_transformer=True)
    
    print(f"Epoch {epoch}/{num_epochs}: "
          f"Train accuracy: {train_acc * 100:.2f}% | "
          f"Val accuracy: {val_acc * 100:.2f}% | "
          f"Train loss: {train_loss:.4f} | "
          f"Val loss: {val_loss:.4f} | "
          f"learning rate: {learning_rate:.6f} | ")
          # f"Precision: {val_metrics['precision'] * 100:.2f}% | Recall: {val_metrics['recall'] * 100:.2f}% | F1: {val_metrics['f1'] * 100:.2f}%")
```
